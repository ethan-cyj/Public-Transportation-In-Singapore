```python
combined_df = pd.concat([hdb_centroid_pair_df, private_centroid_pair_df], axis=0)
```

```python
combined_df.shape
```

```python
combined_df['steepness'] = abs(combined_df['steepness'])

combined_df['time_difference'] = -combined_df['time_difference'] #Reflect time savings as a positive number
```

```python
def calculate_weighted_score(dataframe,row1,row2,row3,row4):
    S = row1 * dataframe["time_difference"] + row2 * dataframe["distance"] + row3 * dataframe["suitability"] + row4 * dataframe["steepness"]
    return S
```

```python
#combined_df['Weighted_Score'] = calculate_weighted_score(combined_df, 0.4, -0.2, 0.8, -0.1)
```

```python
combined_df.head(5)
```

```python
agg_dict = {
    'time_difference': 'mean',
    'distance': 'mean',
    'steepness': 'mean',
    'suitability': 'mean'
}

df = combined_df.groupby('MRT.Name').agg(agg_dict)
df = df.reset_index()

df['Weighted_Score'] = calculate_weighted_score(df, 0.4, -0.2, 0.8, -0.1)
```

```python
mrt_stations_df = pd.read_csv(r"..\data\mrt_station_final.csv",usecols = [1,2,3])
mrt_stations_df.sort_values(by='MRT.Name', inplace=True)

mrt_stations_df
```

```python
mrt_stations_df = pd.merge(mrt_stations_df, df, left_on='MRT.Name',right_on = "MRT.Name",how = "left")

mrt_stations_df
```

```python
ranking = df.sort_values(by='Weighted_Score',ascending = False)
ranking = ranking.reset_index(drop = True)
ranking['rank'] = ranking.index

ranking 
```

```python
ranking.to_csv("../../data/Cluster_data/mrt_ranking.csv", index = False)
```
